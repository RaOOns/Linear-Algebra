### **1. 3 Basic Properties**

#### **1) $det\;I = 1$**

#### **2) $det\;A$ change the sign (부호)**

- **Pivoting을 할 때마다 부호가 변화**

  - 홀수번 발생 : 부호 반대

  - $\begin{vmatrix} a&b \\c&d\end{vmatrix} = ad-bc \quad \begin{vmatrix} c&d \\a&b\end{vmatrix} = cb-ad = -(ad-bc)$

#### **3) $det\;A$ depends linearly on the first row**


- $\quad \begin{vmatrix} a+a'&b+b' \\c&d\end{vmatrix} = \begin{vmatrix} a&b \\c&d\end{vmatrix} + \begin{vmatrix} a'&b' \\c&d\end{vmatrix} \\ \quad \\ \quad (a+a')d - (b+b')c = (ad-bc) + (a'd-b'c)$


- example : $\begin{vmatrix} a&b&c \\d&e&f \\ g&h&i \end{vmatrix} = \begin{vmatrix} a&0&0 \\d&e&f \\ g&h&i \end{vmatrix} + \begin{vmatrix} 0&b&0 \\d&e&f \\ g&h&i \end{vmatrix} + \begin{vmatrix} 0&0&c \\d&e&f \\ g&h&i \end{vmatrix}$


- $\begin{vmatrix} ta&tb \\c&d\end{vmatrix} = t\begin{vmatrix} a&b \\c&d\end{vmatrix}$

  - $\color{red} But \quad \begin{vmatrix} ta&tb \\tc&td\end{vmatrix} = det\;(tA) \begin{cases} \ne t\;det\;A \\ = t^2\;det\;A\end{cases}$

  - $det (tA_{n\times n}) = t^n det(A)$

  - $t\begin{vmatrix} a&b \\ tc&td\end{vmatrix} = -t\begin{vmatrix} tc&td \\ a&b\end{vmatrix} = -t^2\begin{vmatrix} c&d \\ a&b\end{vmatrix} = t^2\begin{vmatrix} a&b \\ c&d \end{vmatrix}$

#### **(4) If two rows are equal, $det(A) = 0$**

   - **Proof)** 
    - $det(A) = -det(A')$　from　$A'$ : row exchange of the same two rows.

    - $a_1^T = a_2^T$

  - $A' = \begin{bmatrix} a_2^T \\ a_1^T \\ a_3^T \end{bmatrix} = A \quad \to \quad det(A) = det(A') \quad$ from $\quad a_1^T = a_2^T$

  - **$det(A) = -det(A')\;$ 이면서 $\;det(A) = det(A')\;$인 경우는 $\;det(A) = 0$ 뿐이다.**

#### **(5) Row operations do not change the determinant**

- **Row operations** : In Gauss Elimination(= G.E), subtracting a multiple of one row from another row


- $det \begin{vmatrix} row1 \\ row2 \\ row3 \end{vmatrix} \;= \;det \begin{vmatrix} \color{Blue} {row1'}\ \\ row2 \\ row3 \end{vmatrix}$


- $\begin{bmatrix} a&b \\ c&d \end{bmatrix} \quad - (row1 - l\cdot row2) \to \quad \begin{bmatrix} a-lc&b-lc \\ c&d \end{bmatrix}$

  - $\begin{vmatrix} a&b \\ c&d \end{vmatrix} + \begin{vmatrix} -lc&-ld \\ c&d \end{vmatrix} \quad \to \quad \begin{vmatrix} a&b \\ c&d \end{vmatrix} - l\begin{vmatrix} c&d \\ c&d \end{vmatrix}\quad =\quad \begin{vmatrix} a&b \\ c&d \end{vmatrix} + 0$


#### **(6) If $A$ has a zero-row, $det(A)$ = 0**

#### **(7) If $A$ is triangular, $det(A)$ = product of diagonal entries**

- $\begin{bmatrix} a&b \\ 0&d \end{bmatrix} = ad \quad \begin{bmatrix} a&0 \\ c&d \end{bmatrix} = ad$


- $U = \begin{bmatrix} d_1&*&\cdots&* \\ 0&d_2&\cdots&* \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&d_n \end{bmatrix} \quad \to \quad Diag$

#### **(8) $A$ : Singular / Non-Sigular**

- $A$ : Singular $\quad \to \quad det(A) = 0\quad \to \quad A^{-1} \;(X)$


- $A$ : Non-Singular $\quad \to \quad det(A) \ne 0\quad \to \quad A^{-1} \;(O)$

#### **(9) $det(AB) \; = \; det(A)\cdot det(B)$**

- $d(A) \triangleq {det(AB) \over det(B)} = det(A)$


- $d(A)$가 $det(A)$인지는 모르지만 **Basic 3 properties**로 증명
  - $B = I$

  - **row exchange → Sign change**

#### **(10) $det(A^T) \; = \; det(A)$**

- $A = LDU$


- $\begin{bmatrix} 1&0&\cdots&0 \\ l_{21}&1&\cdots&0 \\ \vdots&\vdots&\ddots&\vdots \\ l_{n1}&l_{n2}&\cdots&1 \end{bmatrix} \begin{bmatrix} d_1&0&\cdots&0 \\ 0&d_2&\cdots&0 \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&d_n \end{bmatrix} \begin{bmatrix} 1&*&\cdots&* \\ 0&1&\cdots&* \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&1 \end{bmatrix}$


- $det(A) \;=\; det(L) det(D) det(U) \;\; =\;\; 1\;det(D)\;1$


- $A^T \; = \; U^TD^TL^T \; = \; U^TDL^T$


- $det(A^T) \;=\; 1\;det(D)\;1$

## **III. Determinant formula**

### **1. 3-1**

- **$det(A)$ = Product of all pivots from Gauss Elimination (= $\color{red} {no\;\; pivoting}$)**

  - $A = LDU$

  - $det(A)\;\; =\;\; det(L)\; det(D)\; det(U)\;\; =\;\; 1\cdot\prod_{i=1} ^np_i \cdot 1$


- **PA = LDU** with some pivoting

  - $det(P) = \pm 1$

  - $det(A) = \pm (\; \prod_{i=1}^n p_i \;)$

  - Example
  
  - $\begin{bmatrix} 2&-1&0&0&\cdots&0 \\ -1&2&-1&0&\cdots&0 \\ 0&-1&2&-1&\cdots&0 \\ \vdots&\ddots&\ddots&\ddots&\ddots&\vdots \\ 0&0&0&\cdots&-1&2 \end{bmatrix} \; \to \; \begin{bmatrix} 2&-1&0&0&\cdots&0 \\ 0&{3\over 2}&-1&0&\cdots&0 \\ 0&-1&2&-1&\cdots&0 \\ \vdots&\ddots&\ddots&\ddots&\ddots&\vdots \\ 0&0&0&\cdots&-1&2 \end{bmatrix} \; \to \; \begin{bmatrix} 2&-1&0&0&\cdots&0 \\ 0&{3\over 2}&-1&0&\cdots&0 \\ 0&0&{4\over 3}&-1&\cdots&0 \\ \vdots&\ddots&\ddots&\ddots&\ddots&\vdots \\ 0&0&0&\cdots&0&{(n+1)\over n} \end{bmatrix}$

  - **Pivot : $2, {3\over 2}, {4\over 3},{5\over 4},\cdots, {n+1\over n}$**

  - $det(A_n) \;\;= \;\;\prod \;Pivot \;\;=\;\; n+1$

### **3. Big formula**

- $det(A_{3 \times 3}) = a_{11}a_{22}a_{33} + (-1)a_{11}a_{23}a_{32} + \cdots $


- $\sum_{(\alpha, \beta, \gamma)} a_{1\alpha} a_{2\beta} a_{3\gamma} (det (P_{\alpha, \beta, \gamma}))$

  - **$(\alpha, \beta, \gamma)$의 경우의 수 = $3!$**


- $\color{red} {Big \;\; Formula}$

  - $\sum_{(\alpha, \beta, \gamma, \cdots, \nu)} a_{1\alpha} a_{2\beta} a_{3\gamma} \cdots a_{n\nu} \;\; (det (P_{\alpha, \beta, \gamma, \cdots, \nu}))$

  - **경우의 수 = $n!$**

  - $\begin{vmatrix} 1&2&3 \\ 4&5&6 \\ 7&8&9 \end{vmatrix} \;\;= \;\; \begin{vmatrix} 1&0&0 \\ 0&1&0 \\ 0&0&1 \end{vmatrix}_{(1)} (1 \times 5 \times 9) \;+\; \begin{vmatrix} 1&0&0 \\ 0&0&1 \\ 1&0&0 \end{vmatrix}_{(2)} (1 \times 6 \times 8) \;+\; \cdots \;+\; \begin{vmatrix} 0&0&1 \\ 1&0&0 \\ 0&1&0 \end{vmatrix}_{(3)} (3 \times 4 \times 8)$

  - $(1)\;:\; +\;\;/ \;\; (2), (3) \;:\; - \;\; $ **from pivoting**


- **Consider all the terms involving $a_{11}$**

  - $a_{11}a_{22}a_{33}\;(det(P_1)\;\;+\;\;a_{11}a_{23}a_{32}\;(det(P_2))) \\ = a_{11}(a_{22}a_{33}\;det(P_1)\;\;+\;\;a_{23}a_{32}\;det(P_2))\\ = a_{11}\cdot \mathbf{Cofactor}\;(\color{red}{C_{11}})$

  - $a_{12}\cdot Cofactor\;(\color{green}{C_{12}})$

  - $a_{13}\cdot Cofactor\;(\color{green}{C_{13}})$

- $C_{ij} \;=\; (-1)^{i+j}\;\;det(M_{ij}) $
  - $M_{ij} \; :$ **Minor Matrix  =  i행과 j열을 제외한 sub matrix**

### **4. Determinant of $A_{n\times n}$**

#### **1) Gauss Elimination $A \to U$**

- **$det(A)$ = Product of pivots = $(-1)^{number\; of \; row \; exchange} \prod_{i=1}^n P_i$**

#### **2) Big Formula**

- $det(A) \;\; = \;\; \sum_{(\alpha, \beta, \gamma, \cdots, \nu)} a_{1\alpha} a_{2\beta} a_{3\gamma} \cdots a_{n\nu} \;\; (det (P_{\alpha, \beta, \gamma, \cdots, \nu}))$

- $\alpha, \beta, \gamma, \cdots, \nu$ = **peprmutation (총 n! 개)**

#### **3) cofactor**

- **Focus on $\color{red}{One}$ row**

- $det(A) = \sum_{j=1}^n a_{ij} C_{ij}$

- $C_{ij} \;=\; (-1)^{i+j}\;\;det(M_{ij}) $
  - $M_{ij} \; :$ **Minor Matrix**

- $det(A) = a_{11}C_{11}+a_{12}C_{12}+\cdots+a_{1n}C_{1n} \\  \quad \quad \quad = a_{21}C_{21}+a_{22}C_{22}+\cdots+a_{2n}C_{2n}$

## **VI. Determinant Applications**

### **1. $\mathbf{C} \;:\;$ cofactor matrix**

- $\mathbf{C} = \begin{bmatrix} C_{11}&C_{12}&C_{13}&\cdots&C_{1n} \\ C_{21}&\ddots&\ddots&\vdots&\vdots \\ \vdots&\vdots&\ddots&\ddots&\vdots \\ C_{n1}&C_{n2}&C_{n3}&\cdots&C_{nn} \\ \end{bmatrix}$


- $A^{-1}  = {\mathbf{C} \over det(A)}\quad \quad \quad det(A) = (det A)I = A\mathbf{C}^T$

  - $A\mathbf{C}^T \; = \begin{bmatrix} a_{11}&a_{12}&a_{13}&\cdots&a_{1n} \\ a_{21}&\ddots&\ddots&\vdots&\vdots \\ \vdots&\vdots&\ddots&\ddots&\vdots \\ a_{n1}&a_{n2}&a_{n3}&\cdots&a_{nn} \\ \end{bmatrix} \begin{bmatrix} C_{11}&C_{21}&C_{31}&\cdots&C_{n1} \\ C_{12}&\ddots&\ddots&\vdots&\vdots \\ \vdots&\vdots&\ddots&\ddots&\vdots \\ C_{1n}&C_{2n}&C_{3n}&\cdots&C_{nn} \\ \end{bmatrix} \\ \quad \quad \;= \begin{bmatrix} det(A)&0&0&\cdots&0 \\ 0&\det(A)&\cdots&\cdots&\vdots \\ \vdots&\vdots&\ddots&\ddots&\vdots \\ 0&0&0&\cdots&det(A) \\ \end{bmatrix}$

### **2. Cramer's Rule**

- $AX = b$


- $X = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} \;\;=\;\;A^{-1}b \;\;=\;\;A^{-1}\begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n \end{bmatrix}$


- $x_i \;\; = \;\; {det(B_i) \over det(A)} \quad \to \quad \color{red}{Cramer's \;\; Rule}$


- $A^{-1}b = {\mathbf{C}^Tb \over det(A)} \quad\to\quad  \begin{bmatrix} C_{11}&C_{21}&C_{31}&\cdots&C_{n1} \\ C_{12}&\ddots&\ddots&\vdots&\vdots \\ \vdots&\vdots&\ddots&\ddots&\vdots \\ C_{1n}&C_{2n}&C_{3n}&\cdots&C_{nn} \\ \end{bmatrix}\begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n \end{bmatrix} = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$

  - **(1행)** $\;\; b_1C_{11}+b_2C_{21}+b_3C_{31}+cdots+b_nC_{n1}\quad = \quad \color{red}{x_1}$

  - **From**$\quad det(A) = det(A^T)\quad and \quad a_{11}C_{11}+a_{12}C_{12}+\cdots = a_{11}C_{11}+a_{21}C_{21}+\cdots$

  

　　　　　image.png

- **Example**

  - $\begin{cases} x + y = 4 \\ 2x - y = -1 \end{cases} \quad\to\quad \begin{bmatrix} 1&1 \\ 2&-1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 4 \\ -1 \end{bmatrix}$

  - $det(A) = -3$

  - $det(B_1) \;=\; \begin{vmatrix} 4&1 \\ -1&-1 \end{vmatrix} \;=\; -3$
  
  - $det(B_2) \;=\; \begin{vmatrix} 1&4 \\ 2&-1 \end{vmatrix} \;=\; -9$

  - $x = {-3 \over -3} = 1\quad, \quad y = {-9 \over -3} = 3$

### **3. Volume of Box**

- $A = \begin{bmatrix} \\ a_1&a_2&\cdots&a_n \\ &&& \end{bmatrix}$


- $det(A) = Volume$

　　image.png

- **for right-angled(= 수직) box**

  - $||a_1|| = l_1, \; ||a_2|| = l_2, \; \cdots$

  - $a_i \bot a_j$

  - $A = \begin{bmatrix} \\ a_1&a_2&\cdots&a_n \\ &&& \end{bmatrix}$

  - $det(A^TA) = det\left(  \begin{bmatrix} a_1^T \\ a_2^T \\ \vdots \\ a_n^T \end{bmatrix} \begin{bmatrix} \\ a_1&a_2&\cdots&a_n \\ &&& \end{bmatrix} \right) \;\;= \;\; det\left(  \begin{bmatrix} ||a_1||^2 &0&\dots&0 \\0&||a_2||^2&\cdots&0 \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&||a_n||^2 \end{bmatrix} \right)$

  - $det(A^TA)\;\; =\;\; (det(A))^2\;\; =\;\; l_1^2\;l_2^2\;\cdots\;l_n^2 $

  - **서로 수직이면**$\quad det(A) = l_1\;l_2\;\cdots \; l_n$
