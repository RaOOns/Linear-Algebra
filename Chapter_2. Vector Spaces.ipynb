## **I. Vector spaces and Subspaces**
---

### **1. A real vector Space $R^n$**

- $R^n$　:　n-dimensional


- $x+y = y+x$
- $x + (y+z) = (x+y) + z$
- There is a unique **"zero vector"** such that $x+0 = x, ∀ x$
- For each $x$, There is a unique vector $-x$, such that $-x+x = 0$
- $1 \cdot x = x$
- $(C_1C_2)x = C_1(C_2x)$
- $C(x+y) = Cx + Cy$
- $(C_1 + C_2)x = C_1x + C_2x$　　　$x,y$ : vector ,　 $C_1, C_2, C_3$ : scalar



- $x$ and $y$ should be closed under addition and scalar multiplication　→　**"Space"**

- Example
  - $R^{∞}$ space
  - $m \times n$ matrix　→　$R^{m \times n}$
    - 최대 2차 다항함수　:　$ax^2 + bx + c \in R^3$　 $(a\ b\ c)^T = (3\times1) = 3$
  - funtion f(x)　→　$R^{∞}$　→　$f(x) = ae^x = 1 + x + {x^2 \over 2!} + {x^3 \over 3!} \cdots $ (from Taylor series). 　　$(1\ 1\ 1/2!\ 1/3!\ \cdots) = R^{∞}$


### **2. Subspace**

  - non empty subset that satisties the requirements for a vector space.
  
  - $x+y \in subspace$,　for $∀x, y \in subspace$
  - $cx \in subspace$,　for $∀x \in subspace, ∀c$

  - For example
    - $R^3$ vectors on a plane which passes through the origin.
    - the origin　→　smallest subspace.
    - $R^2$vectors on aline which passes through the origin.
    - Lower triangular matrices.

    - (Upper or) Lower Triangle Matrix in $R^{n^2}$
    $L_{n\times n}　|　c_1L_1 + c_2L_2　\in　L_{n\times n}　\in　A_{n\times n}　\in　R^{n^2}$

### **3. The Column Sace of A**


- **MY.** $A$행렬의 각 열을 의미하는 열 벡터로 $A$의 역행렬인 $A^{-1}$이 존재하는지를 판단하는데 사용된다.
　　$AX = b　→　X = A^{-1}b$이므로 $A^{-1}$의 존재 유무는 $X$ 벡터의 존재(해의 유무)를 결정한다.


- $C(A)$

- Column Space　:　contains all linear combinations of the columns of matrix $A^{mn}$.
  - $Ax = b$　→　$\begin{bmatrix}a_1&a_2&\cdots&a_n \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} = \begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n\end{bmatrix}$

  - $a_n$ : column vectors.

  - $x_1a_1 + x_2a_2 + \cdots + x_na_n = b$　→　linear combinations of column vectors equals $b$
  - $C(A) = [X | X = \sum_{i=1}^n C_ia_i]$
  - $\begin{cases} if　　b \in C(A)　\rightarrow　there\ are\ solutions. \\ elif　b \notin C(A)　\rightarrow　No\ solution.\end{cases}$

\  
- $Ax = b$ is solvable iff $b$ can be expressed as a combination of the columns of $A$, Then $b$ is in the column space.
  ex) $u\begin{bmatrix}1 \\ 5 \\ 2 \end{bmatrix} + v\begin{bmatrix}0 \\4 \\ 4 \end{bmatrix} = \begin{bmatrix} b_1 \\ b_2 \\ b_3\end{bmatrix}$

  image.png

- vector $b$ should be closed to the subpace spanned by cloumn vectors

- Column Space $C(A)$ **Conditions**
  - if $b$ and $b'$ lie in the column space 　→　$A(x+x') = b+b'$　→　closed under addition.
  - if $b$ lies in the column space 　→　$A(cx) = cb$　→　closed under scalar multiplication.

- Example
  - for any non-singular 5X5 matrix　→　$R^5$ (column) space　→　5 pivots in Gauss elimination.
  - Zero matrix　→　0-dimensional space.

- $Ax = b$ 　　$A_{n \times n}$
  - $X = A^{-1}b$　→　A의 Inverse가 존재하면 $X = unique$하게 존재. 
  　　　　　　　　 if $A^{-1}$ exists, alway $b \in C(A)$　→　$C(A)$ whole space.

  - Ex. $A = \begin{bmatrix} 1&-2&1 \\ 2&1&1 \\ 0&0&0 \end{bmatrix}$
  　→　아무리 선형결합을 하더라도 $z$ = zero-vector. 즉, 3차원 공간이어도 2차원뿐으로 
  　　　이 vectors 들은 3차원 공간을 전부 설명하지 못한다. **But Inverse가 존재한다면 그 n차원을 모두 설명가능.** 

  - Whole space is constructed $a_1, \cdots, a_n$ by linear combinations.　→　**span**

### **4. Null Space of A, N(A)**


- **MY.** 해집합으로 $AX = 0$을 만족하는 모든 $X$ 벡터를 의미하고 있으며,
　　해가 없거나, 무수히 많은 경우로 나뉜다. 


- Set of Vectors such that $AX = 0$
  - $N(A) = [X | AX = 0]$
  
  - Conditions
    - closed under addtion.　→　for $Ax_1 = 0 $, $AX_2 = 0$ $X_1, X_2 \in N(A)$ then, $A(X_1 + X_2) = 0$ 

    - closed under scalar multiplication.　→　for $AX_1 = 0$ for any $C$,　$CX \in N(A)$ then $A(CX) = 0$


- Example.

  $\begin{cases} u　+　w = 0 \\ 5u + 4v +9w = 0 \\ 2u+4v+6w = 0\end{cases}$　→　$\begin{cases} u+w = 0 \\ v+w = 0\end{cases}$　→　$if　w = c \\ then,　u = v = -c$　→　$\begin{bmatrix} u \\ v \\ w \end{bmatrix} = c \begin{bmatrix} -1 \\ -1 \\ 1 \end{bmatrix}$　→　**공간에서 원점을 지나는 직선**
  
  　→　$AX = 0$ 꼴로도 표현 가능. $\begin{bmatrix} 1&0&1 \\ 5&4&9 \\ 2&4&6 \end{bmatrix}\begin{bmatrix} u \\ v \\ w \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}$

  　→　여기서 해인, $c \begin{bmatrix} -1 \\ -1 \\ 1 \end{bmatrix}$는 $N(A)$의 조건인 $c_1 \begin{bmatrix} -1 \\ -1 \\ 1 \end{bmatrix} + c_2 \begin{bmatrix} -1 \\ -1 \\ 1 \end{bmatrix} = (c_1 + c_2) \begin{bmatrix} -1 \\ -1 \\ 1 \end{bmatrix}$를 만족하므로, 
  　　　$\in N(A)$이며 이는 곧, **Infinity many solution**이다.

## **III. Linear Independence(선형 독립)**
---


### **1. Linear Independent**

- $c_1v_1 + c_2v_2 + \cdots + c_nv_n = 0$에서 모든 $v$들이 선형독립인 경우는
Only $c_1 = c_2 = \cdots = c_n = 0$인 경우.

- Any vector $v_k$ **can not** be expressed by the other vectors.


- Example 1

  $c_1 \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} + c_2 \begin{bmatrix}  \\ 1 \\ 0 \end{bmatrix} + c_3 \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix} = \begin{bmatrix} c_1 \\ c_2 \\ c_3 \end{bmatrix} =  \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}$의 경우 $c_1 = c_2 = c_3 = 0$일 때만 가능.

- Example 2
  
  $A = \begin{bmatrix} 3&4&2 \\ 0&1&5 \\ 0&0&2 \end{bmatrix}$　→　Column vectors


- **If G.E of $A$ generates m non-zero rows**

  $\begin{bmatrix} &&&& \\ &&&& \\ 0&0&0&0&0\end{bmatrix}$　→　 **row가 non-zero인 개수 = Independant 개수　→　independent column vectors in A**


### **2. Rank of $A$**
  - = **# of Independant column vectors**
  - = **# of Independant row vectors**
  - = **# of Pivot in G.E**
  - = **Dim of $C(A)$**

  - Example $\begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix}$　→　아무리 결합해도 $z=0 \;$ 즉, $x, y$평면(2차원)
  - Example $\begin{bmatrix} 1&3&3&2 \\ 2&6&9&7 \\ -1&-3&3&4 \end{bmatrix}$　→　1열 x 3 = 3열 / 3열 - 1열 = 4열


  - **Dimension of vector space 
  = # of linearly indep vectors**

 ### **3. Spanning**

  - **MY.** 어떠한 벡터($v_1, v_2, v_3$)들의 **조합**들로 만들어질 수 있는 **모든** 벡터 공간을 의미.

  - All linear Combinations of vectors {$v_1, v_2, v_3$} constructed a vector space.
  = {$v_1, v_2, v_3$} **span** vector space.

  - **공간(or 영역)을 Span하는 vector는 Unique하지 않다.**

  - If given linearly indep vectors　→　linear combination　→　unique


  - **Example**
    
 　　 $\begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix}$　→　$xy$평면만 **span**가능　　$\begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix}$　→　$xy$평면만 **span**가능　　$\begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix} \begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix}$　→　$xy$평면만 **span**가능

　　　　　 $b = \begin{bmatrix} 2 \\ 2 \\ 0 \end{bmatrix}$인 경우에 대해　 왼쪽은 $c_1 = c_2 = 2$　 중간은 $c_1 = 0, c_2 = 2$로 **해가 1개** 뿐이지만, 
　　　　　 오른쪽은 $\begin{cases}c_1 = c_2 = 2, c_3 = 0 \\ c_1 = c_2 = 0, c_3 = 2 \\ c_1 = c_2 = c_3= 1\end{cases} $로 **여러 개의 해**가 존재한다.



### **4. Basis (vector)**

　　**MY. 공간을 구성하는데 필요한 최소 (선형 독립인) Vectors.
　　　　해가 1개인지 여러 개인지를 파악하는데 사용되는 것이 Basis (vector) 이다.**


- 공간을 span하는 vector는 몇개던지 상관 X

- Basis vector는 **"# of minimum linearity indepedent vectors to span the vector span"**

- **Maximum # of linearly Indep vectors**

- **Linear combination is unique from basis**

- **# Basis is not unique for a vector space**
  - $ v_1 = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} v_2 = \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix} v_3 = \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix} \subset R^3$
  - $ v_1 = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} v_2 = \begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix} v_3 = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix} \subset R^3$
  - 두 경우 모두 $R^3$ 공간의 기저 벡터들이지만 값이 다르다.


- Linearly Indep　→　수직관계　→　**$v_i \cdot v_j = 0, \quad i \ne j$**　→　(회귀식) Corr = 0$\;\;$ 즉, 다중공선성이 없음을 의미.

  - **주의!!**　:　**직교이면 선형독립이다.
　　　　　그러나 선형독립이면 직교인 것은 아니다!!!**

  - $v_1, v_2, \cdots, v_n$ : Basis vectors 　→　$b = c_1v_1+\cdots+c_nv_n$　→　$v_k \cdot b = v_k \cdot (c_1v_1 + \cdots + c_nv_n) = c_k||v_k||$　→　$\begin{bmatrix}v_1&v_2&\cdots&v_n \end{bmatrix}\begin{bmatrix} c_1 \\ \vdots \\ c_n\end{bmatrix} = b$　→　$AX = b$ 형태.

## **IV. Linear Transformation**
---

### **1. linear transformation**

- $AX = b$를 바라보는 관점
  1. System equation　→　$X$ = solution

  2. Linear Combination of Column vectors　→　$X$ = Scalar coef

  3. **Linear Transformation.**


image.png

- Example
  - $\begin{bmatrix} c&0 \\ 0&c \end{bmatrix}$　→　Stretching  or  Contracting 변환

  - $\begin{bmatrix} 0&-1 \\ 1&0 \end{bmatrix}$　→　90º rotation 변환

  - $\begin{bmatrix} c&0 \\ 0&c \end{bmatrix}$　→　$X ↔ y$ 좌표값 교환,　$y = x$ reflection (대칭)


- Differentiation (${d \over dt}$) of polynomial.
  - $X(t) = a_0 + a_1t + a_2t^2 + \cdots +a_nt^n$
  - $y(t) = {d \over dt}X(t) = b_0 + b_1t + b_2t^2 + \cdots + b_{n-1}t^{n-1}$
  - $R^{n+1}　-{d \over dt}→　R^{n}$
   $\begin{bmatrix} a_0\\a_1\\ \vdots \\a_n \end{bmatrix}　-{d \over dt}→　\begin{bmatrix} b_0\\b_1\\ \vdots \\b_{n-1} \end{bmatrix}$


- Integration ($\int_{-∞}^t X(t)dt$) of polynomial.
  - $X(t) = a_0 + a_1t + a_2t^2 + \cdots +a_nt^n$
  - $y(t) = {d \over dt}X(t) = b_0 + b_1t + b_2t^2 + \cdots + b_{n-1}t^{n-1}$
  - $R^{n+1}　→　R^{n+2}$
   $\begin{bmatrix} a_0\\a_1\\ \vdots \\a_n \end{bmatrix}　→　\begin{bmatrix} b_0\\b_1\\ \vdots \\b_{n+1} \end{bmatrix}$


- Multiplication by a polynomial
  - $x(t)p(t) = y(t)$
  - It is linear　from　$(a_1x_1(t) - a_2x_2(t))p(t) = a_1x_1(t)p(t) + a_2x_2(t)p(t)$
  - if　$p(t) = p_0 + p_1t + p_2t^2$
  - $x(t) = a_0 + a_1t + a_2t^2 + \cdots +a_nt^n$

### **2. How to find $A$?**

→　If we know the $AX$ for the elementary basis vectors.
$X_1 = \begin{bmatrix} 1\\0\\0\\0\\ \vdots \\0 \end{bmatrix} → a_1,　X_2 = \begin{bmatrix} 0\\1\\0\\0\\ \vdots \\0 \end{bmatrix} → a_2,　X_3 = \begin{bmatrix} 0\\0\\1\\0\\ \vdots \\0 \end{bmatrix} → a_3,　\cdots, X_n = \begin{bmatrix} 0\\0\\0\\0\\ \vdots \\1 \end{bmatrix} → a_n$
$⇒ A = \begin{bmatrix} \\ a_1&a_2&\cdots&a_n\\ \\ \end{bmatrix}$



- **Example**
$X_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix} → \begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix}　X_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix} → \begin{bmatrix} 4 \\ 6 \\ 8 \end{bmatrix}$
$⇒ A = \begin{bmatrix} 2&4 \\ 3&6 \\ 4&8 \end{bmatrix}$
$∴ A\begin{bmatrix} -1 \\ 2 \end{bmatrix}  = \begin{bmatrix} 6 \\ 9 \\ 12 \end{bmatrix}$ 

 **for different basis vectors (MY. 1과0으로 구성된 기저벡터가 아닌 경우?)**
$X_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix} → \begin{bmatrix} 6 \\ 9 \\ 12 \end{bmatrix}　X_2 = \begin{bmatrix} 2 \\ -1 \end{bmatrix} → \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}$
$⇒ A = \begin{bmatrix} 6&0 \\ 9&0 \\ 12&0 \end{bmatrix}$
$∴ A\begin{bmatrix} -1 \\ 2 \end{bmatrix}  = \begin{bmatrix} -6 \\ -9 \\ -12 \end{bmatrix}$
 **→　잘못된 결과가 나온다.**
 **→　이때의 풀이법**
 　　　$\begin{bmatrix} -1 \\ 2 \end{bmatrix} = X_1 + X_2$
 　　$A\begin{bmatrix} -1 \\ 2 \end{bmatrix} = AX_1 + AX_2 = \begin{bmatrix} 6 \\ 9 \\12 \end{bmatrix}$
