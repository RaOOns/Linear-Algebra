### **1.Orthogonal vector**

- **Why orthogonal vectors? (왜 직교 벡터여야지?)**
  - Independent(basis)　→　**직교하면 서로 독립**
  - Easy calculation in linear combination　→　**선형 결합 계산이 쉽다**


- Length of vectors $||X||$
  - $||X||^2 = X^TX = \sum_{i=1}^n x_i^2$


　　　　　image.png

 - By Pythagoras Theorem
  $||y-X||^2 = ||X||^2 + ||y||^2$
  　　→　$(y-x)^T(y-x) = X^TX + y^Ty$
  　　→　$y^Ty - y^TX - X^Ty +x^TX = X^TX + y^Ty　→　X^Ty = y^TX　→　X^Ty = 0$


- Other vector inner product $X^Ty$
  - $X^Ty = 0　→　angle = 90$º
  
  - $X^Ty < 0　→　angle > 90$º
  
  - $X^Ty > 0　→　angle < 90$º


- **If non-zero vectors $v_1, v_2, \cdots, v_n$ are othogonal　→　$v_i^Tv_j = 0　(i \ne j),　||v_i|| \ne 0$
→　then, the vectirs are linearly Indep!**
  - (proof) $c_1v_1 + c_2v_2 + \cdots + c_nv_n = 0$인 경우가 **오직** $c_1=c_2=\cdots=c_n=0$일 뿐이면 
$v_i^T(c_1v_1 + c_2v_2 + \cdots + c_nv_n) = 0$은　$c_i||v_i||^2 = 0　for　∀ i$
→　$c_i = 0$이면,　$v_i$는 Independent & Non-Zero 이고,　orthogonal vectors　→　basis vectors가 된다.


- $X = \sum_{i=1}^n c_iv_i　→　v_i^TX = c_i||v_i||^2$　**Orthogonal Normal 이면 $||v_i|| = 1$**


- linearly Independent vectors
  - $a_1, a_2, \cdots, a_n$
  - $X = \sum_{i=1}^n c_ia_i$　→　$c_i$ 구하기　→　$\begin{bmatrix} a_1&a_2&\cdots&a_n\end{bmatrix} \begin{bmatrix}c_1 \\ \vdots \\ c_n \end{bmatrix} = \begin{bmatrix} X \end{bmatrix}$　→　**Orthogonal Normal**이면,　$v_i^TX = c_i||v_i||^2$ 사용.

## **IV. Gram-Schmitt Orthogonalization**
---

image.png

1.　$a　→　{a \over ||a||} = q_1$

2.　project　$b$　onto　$q_1$　→　${q_1^Tb \over q_1^Tq_1}q_1　　from　q_1^Tq_1 = 1$

3.　$b - (q_1^Tb)q_1　\bot　q_1$　→　${b - (q_1^Tb)q_1　\over ||b - (q_1^Tb)q_1||} = q_2$

4.　$b = (q_1^Tb)q_1 + (q_2^Tb)q_2$

image.png

- $a_1, a_2, \cdots$
  - 1) $q_1 = {a \over ||a||}$

  - 2) $a_j - \sum_{i=1}^{j-1} (q_i^Ta_j)q_i　=　A_j$

  - 3) ${A_j \over ||A_j||} = q_j$

  - $X = \sum_{i=1}^{j} (q_i^T X)q_i$

- **Example**
  - $a_1 = \begin{bmatrix} 1\\0\\1 \end{bmatrix},\;\; a_2 = \begin{bmatrix} 1\\0\\0 \end{bmatrix},\;\; a_3 = \begin{bmatrix} 2\\1\\0 \end{bmatrix}$

  - $q_1  = {a_1 \over ||a_1||} = {1 \over \sqrt{2}} \begin{bmatrix} 1\\0\\1 \end{bmatrix}$

  - $q_2 = a_2 - (q_1^Ta_2)q_1 = \begin{bmatrix} 1\\0\\0 \end{bmatrix} - {1 \over \sqrt{2}} \left[{1 \over \sqrt{2}} \begin{bmatrix} 1\\0\\1 \end{bmatrix}\right] = \begin{bmatrix} 1/2\\0\\-1/2\end{bmatrix}　→　q_2 = {1\over\sqrt{2}}\begin{bmatrix}1\\0\\-1 \end{bmatrix}$

  - $q_3 = a_3 - (q_1^Ta_3)q_1 - (q_2^Ta_3)q_2 = \begin{bmatrix} 2\\1\\0 \end{bmatrix} - \sqrt{2}\begin{bmatrix} 1/\sqrt{2}\\0\\1/\sqrt{2} \end{bmatrix} - \sqrt{2}\begin{bmatrix} 1/\sqrt{2}\\0\\1/\sqrt{2} \end{bmatrix} =  \begin{bmatrix} 0\\1\\0 \end{bmatrix}　→ q_3 = \begin{bmatrix} 0\\1\\0 \end{bmatrix}$

## **VI. Orthogoal Basis**
---

- Orthogonal vectors
  - Independent
  - basis vectors


- Let $q_1, q_2, \cdots, q_n$ be **ortho-normal**　$\begin{cases} 0
& i \ne j \\ 1&i = j \end{cases}$
  - $Q = \begin{bmatrix} &&& \\ q_1&q_2&\cdots&q_n \\ &&& \end{bmatrix}$

  - $Q^TQ　→　Q^T = Q^{-1} \quad (Left \quad Inverse)$
  $\begin{bmatrix} -&q_1^T&- \\ -&q_2^T&- \\ &\vdots& \\ -&q_n^T&-\end{bmatrix} \begin{bmatrix} &&& \\ q_1&q_2&\cdots&q_n \\ &&& \end{bmatrix} \quad = \quad I$


### **1. Q examples**

- **Q examples**
  - Rotation Matrix　→　거의 Q-matrix = Rotation Matrix　　$\begin{bmatrix} cos\theta & -sin\theta \\ sin\theta & cos\theta\end{bmatrix}$

  - Permutation Matrix　　$\begin{bmatrix} 0&1&0 \\ 1&0&0 \\ 0&0&1 \end{bmatrix}$ : 1,2행 교환


- **Q transformation preserves the length and angle**
  - $||X||^2,\quad ||QX||^2 = X^TQ^TQX = X^TX \quad from \quad Q^TQ = I$

  - $X^Ty, \quad (QX)^T(Qy) = X^TQQy = X^Ty$

### **2. Projection reduces the length**

image.png

- for $q_1, q_2, \cdots, q_n \in R^n$ (Sqaure sys)
  - $X = \sum{i=1}^n c_i q_i$　→　**$X$를 $q$로 표현**

  - $X = \begin{bmatrix} q_1&\cdots&q_n \end{bmatrix} \begin{bmatrix} c_1\\ \vdots \\c_n \end{bmatrix}$

  - $Q\begin{bmatrix} c_1\\ \vdots \\c_n \end{bmatrix} = \begin{bmatrix} x_1\\ \vdots \\x_n \end{bmatrix}$

  - $\begin{bmatrix} c_1\\ \vdots \\c_n \end{bmatrix} = Q^T \begin{bmatrix} x_1\\ \vdots \\ x_n \end{bmatrix} = \begin{bmatrix} -&q_1^T&-\\ -&\vdots&- \\ -&q_n^T&- \end{bmatrix} X$

  - $c_i = q_i^TX = {q_i^TX \over q_i^Tq_i} \quad from \quad q_i^Tq_i = 1$

  - $X = \sum_{i=1}^n c_iq_i　→　q_i^TX = q_i^T(\sum{i=1}^nc_iq_i) = q_i^T(c_1q_1 + c_2q_2 + \cdots + c_nq_n) = c_i ||q_i||^2 = c_i$


- for $q_1, q_2, \cdots, q_n \in R^n$ (Rectangular Sys, m < n)
  - $Q = [q_1, q_2, \cdots, q_n]$

  - $QX = b 　→　X = (Q^TQ)^{-1}Q^Tb 　→　Q^Tb \quad from \quad Q^TQ = I$

  - $Q$ : **Left-Inverse for retangular sys** 
  $Q = \begin{bmatrix} 1&0 \\ 0&0 \\ 0&1 \end{bmatrix}, \quad Q^T = \begin{bmatrix} 1&0&0 \\ 0&1&0\end{bmatrix}$
  $Q^TQ = \begin{bmatrix}1&0 \\ 0&1 \end{bmatrix}, \quad QQ^T = \begin{bmatrix} 1&0&0 \\ 0&0&0 \\ 0&1&0 \end{bmatrix}$

## **I. Function Space**
---

- Expand vector space to functions.

- $V.S \quad in\quad R^n$　→　$H.S \quad in \quad R^\infty$

- $x(t), y(t) \in H,\quad a \leq t \leq b$
  - $x(t) + y(t) \in H$
  - $\alpha x(t) \in H$

### **1. Range**

- Vectors : $V = (v_1,v_2, \cdots, v_n)^T$

- Funtion space : $ \lim_{\triangle t \to 0} x(t) = (x(a), x(a+\triangle t), x(a+2\triangle t), \cdots, x(b))^T \quad \in R^\infty$

### **2. Inner Product**

- Vector inner product　 :　 $X^Ty = \sum_{i=1}^n x_i y_i$

- Function inner product　 :　 $(x(t), y(t)) = \lim_{\triangle \to 0} \sum_{k=o}^\infty x(a+k\triangle t) y(a+k\triangle t) = \int_a^b x(t) y(t) dt$
  - $||x(t)||^2 = \int_a^bx^2(t) dt$

### **3. Orthogonality**

- $X^Ty = 0 \quad \to \quad (x(t), y(t))\quad =\quad \int_a^b x(t) y(t) dt = 0$

### **4. Basis**

- Basis (vectors)　:　$X = \sum_{i=1}^n a_i v_i $

- Basis (functions)　:　$x(t) = \sum_{i=1}^\infty a_i b_i(t)$　→　**Series라고 부른다!!**

- **Ortho-normal**　:　$\sum (q_i^Ta)q_i \quad \to \quad \sum_{i=1}^infty (q_i(t) x(t))q_i(t)$

### **5. If given a basis**

- If given a basis vectors　→　linear Combination　→　$\{ a_i\}\;\; are\;\; unique$

- If given basis **functions**　→　**Series coefficients are unique**

  - Especially, $\{ cos(nt), sin(mt)\} \quad m, n$ are integer $\quad 0 \leq t \leq 2\pi$
    - $n = m = 0$　→　DC 성분

    - **Frequency (주파수)　=　# of Oscillation(진동) in a period** 

### **6. Orthogonal basis funtion**

- $\int_0^{2\pi} cos n_1t\; cos n_2t \;dt$　=　$\begin{cases} {1\over 2}\int_0^{2\pi} cos(n_1+n_2)t + cos(n_1-n_2)t \; dt = 0 &n_1 \ne n_2 \quad \to\quad Orthogonal \\ {1\over 2}\int_0^{2\pi} (1 + cos 2n_1t) \;dt = \pi &n_1 = n_2  \end{cases}$ 


- $\int_0^{2\pi} cos nt \; sin mt \; dt = 0 \quad \to\quad Orthogonal$


- $\int_0^{2\pi} sin (m_1t) \; sin (m_2t) \; dt \begin{cases} = 0 &nm1 \ne m_2 \quad \to\quad Orthogonal \\  = \pi &m_1 = m_2  \end{cases}$
