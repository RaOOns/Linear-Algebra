### **1. Remark 1**

- **If $\lambda_1, \lambda_2, \lambda_3, \cdots, \lambda_n \;$ are diffenent $\quad$ then, $\;e_1, e_2, \cdots, e_n \;$ are $\;\color{red}{Linearly\;\;Independent}$**

\
- **Proof)** Aussme that $e_2$ is Independent on $e_1\quad\to\quad$ **"모순임을 밝히면 된다."** 

  - "$e_2 = ce_1,\quad c\;\;:$상수"$\;\;$로 표현 가능하며, $c\;=\;0$이어야 **Independent**이다.

  - $(\times A)\quad Ae_2 = cAe_1 \quad\to\quad \lambda_2e_2 \;=\; c\lambda_1e_1\quad\cdots\cdots(1)$

  - $(\times \lambda_2)\quad \lambda_2e_2 = c\lambda_2e_1 \quad\cdots\cdots(2)$

  - (1)과 (2)를 통해 $\;\; c\lambda_2e_1 = c\lambda_1e_1 \quad\to\quad c(\lambda_2-\lambda_1)e_1 \;=\;0$

  - $e_1 \ne 0 \quad $from eigenvector, $\;\;\lambda_1 \ne \lambda_2$ 따라서, $c \;=\;0$

  - **이 방법을 $e_3, e_4, \cdots$에 모두 적용**

  - $e_n \;=\; c_1e_1 + c_2e_2 + \cdots + c_{n-1}e_{n-1}\quad\to\quad c_1 = c_2 = \cdots = c_{n-1} = 0$임을 확인

  - $(\times A)\quad Ae_n \;=\; c_1Ae_1 + c_2Ae_2 + \cdots + c_{n-1}Ae_{n-1} \quad\to\quad \lambda_ne_n \;=\; c_1\lambda_1e_1 + c_2\lambda_2e_2 + \cdots + c_{n-1}\lambda_{n-1}e_{n-1}\cdots(1)$\
  from Eigenvector의 원리

  - $(\times \lambda_n)\quad \lambda_ne_n \;=\; c_1\lambda_ne_1 + c_2\lambda_ne_2 + \cdots + c_{n-1}\lambda_ne_{n-1} \quad\cdots\cdots(2)$

  - (1)과  (2)를 통헤 $\;\; c_1(\lambda_n-\lambda_1)e_1 + c_2(\lambda_n-\lambda_2)e_2 + \cdots + c_{n-1}(\lambda_n-\lambda_{n-1})e_{n-1} \;=\;0$

  - **$e_i \ne 0,\;\; \lambda_i \ne \lambda_j\quad(i\ne j)$ 이므로 $\;\;c_1 = c_2 = \cdots = c_{n-1} = 0$이다.**

- **If $\;\; A^T \; =\; A \quad\to\quad e_1 \;\bot\; e_2 \;\;and\;\; \lambda_1 >> \lambda_2$**

### **1. Fibonazzi Sequence**

\
- $F_{k+2} \;=\; F_{k+1} + F_k \;,\quad k \;=\;0,1,2,\cdots$

- $F_0 = 0,\;\; F_1 = 1,\;\; F_2 = 1,\;\; F_3 = 2,\;\; F_4 = 4, \;\;\cdots$

- $F_{1000} = ?$

\
**(Solution)**
- $F_{k+2} = F_{k+1} + F_k \\ F_{k+1} = F_{k+1}$ $\quad and \quad u_k = \begin{bmatrix} F_{k+1} \\ F_k \end{bmatrix} \quad 이면 \quad u_{k+1} = \begin{bmatrix} 1&1 \\ 1&0 \end{bmatrix} u_k$ 

- 즉, $u_{k+1} = Au_k$ 꼴로 표현 가능
  - $u_1 = Au_0$
  - $u_2 = Au_1 = A^2u_0$
  - $u_k = A^ku_0$

- **여기서 $A^k$를 $\;\;A \;=\; S\Lambda S^{-1}$를 사용하여 $\;A^k = S\Lambda^kS^{-1}$로 표현**

- $A^k = S\Lambda^kS^{-1} \;\;=\;\;\begin{bmatrix} e_1&e_2 \end{bmatrix}\begin{bmatrix} \lambda_1^k&0 \\ 0&\lambda_2^k \end{bmatrix}\begin{bmatrix} S^{-1} \end{bmatrix}$로 표현된다.

- 이로 인해, $u_k  \;\;=\;\;\begin{bmatrix} e_1&e_2 \end{bmatrix}\begin{bmatrix} \lambda_1^k&0 \\ 0&\lambda_2^k \end{bmatrix} \begin{bmatrix} c_1 \\ c_2 \end{bmatrix} \quad\quad from \;\; u_k = A^ku_0 \;\;and \;\; \color{darkblue}{S^{-1}u_0 \;=\; \begin{bmatrix} c_1 \\ c_2 \end{bmatrix}}$

- 다시 정리하면, $\;\; u_k \;\;=\;\; \begin{bmatrix} e_1&e_2 \end{bmatrix}\begin{bmatrix} c_1\lambda_1^k&0 \\ 0&c_2\lambda_2^k \end{bmatrix} \;\;=\;\; \color{red}{c_1\lambda_1^ke_1 + c_2\lambda_2^ke_2}$

- 여기서 $\;\lambda_1, \lambda_2$는 어떻게 구하는가?
  - $|A - \lambda I| \;=\; \begin{vmatrix} 1 - \lambda&1 \\ 1&-\lambda \end{vmatrix} \;=\; 0 \quad\quad from \;\; A = \begin{bmatrix} 1&1 \\ 1&0 \end{bmatrix} \;\; and\;\;\; det(A) = 0$

  - $\lambda^2 - \lambda -1 \;=\;0 \quad\to\quad \color{red}{\lambda_1 = {1+\sqrt{5} \over 2}\;,\; \lambda_2 = {1-\sqrt{5} \over 2}}$


- 이를 통해 $S$를 구하면, $\;\; \begin{bmatrix} 1-\lambda&1 \\ 1&-\lambda \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \;=\; \begin{bmatrix} 0 \\0  \end{bmatrix} \quad\to\quad x_1-\lambda x_2 \;=\; 0 \quad\to\quad \color{red}{S = \begin{bmatrix} \lambda_1&\lambda_2 \\ 1&1  \end{bmatrix}}$

- 따라서 $\;\; S^{-1} = {1 \over \lambda_1 - \lambda_2}\begin{bmatrix} 1&-\lambda_2 \\ -1&\lambda_1  \end{bmatrix}$이고, $\quad S^{-1}u_0 \;=\; {1 \over \sqrt{5}} \begin{bmatrix} 1&-\lambda_2 \\ -1&\lambda_1  \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} \quad\quad from \;\; u_0 = \begin{bmatrix} F_1 \\ F_0 \end{bmatrix}$

\
- $c_1\lambda_1^ke_1 + c_2\lambda_2^ke_2 \;\;=\;\; ({1 \over \sqrt{5}}) ({1+\sqrt{5} \over 2})^k \begin{bmatrix} \lambda_1 \\ 1 \end{bmatrix} + ({1 \over \sqrt{5}}) ({1-\sqrt{5} \over 2})^k \begin{bmatrix} \lambda_2 \\ 1 \end{bmatrix}\quad\quad \lambda_1 = {1+\sqrt{5} \over 2}\;,\; \lambda_2 = {1-\sqrt{5} \over 2}$

\
- **따라서 $\;\; \color{red}{F_k} \;\;=\;\; ({1 \over \sqrt{5}}) ({1+\sqrt{5} \over 2})^k + ({1 \over \sqrt{5}}) ({1-\sqrt{5} \over 2})^k \;\;$이다.**


### **2. Characteristic equation**

\
- $F_{k+1} = F_{k+1} + F_k \quad\to\quad F_{k+1} - F_{k+1} - F_k = 0$

\
- Difference가 2개 이므로, Solution의 형태는 $c_1(\lambda_1)^k + c_2(\lambda_2)^k \;$이다.  $\quad$from 수학적 개념

- $\lambda^k$가 Solution이라 가정하고 대입한 뒤 $\lambda^k$를 제거하면 $\lambda^2 - \lambda - 1 = 0\;\;$식이 나오며 이런식을 **"Characteristic Equation"** 이라 한다.

\
- Example

  - $a_{n+2} - 2 a_{n-1} - a_n = 0$

  - Solution $\;\;=\;\; \lambda^n \;$이라 가정하도 대입

  - $\lambda^{n+2} - 2\lambda^{n+1} - \lambda^n = 0$

  - $\to\quad \lambda^2 - 2\lambda - 1 = 0$

  - $\therefore \lambda = 1 \pm \sqrt{2}$

  - $Solution \;\; a_n \;\;=\;\; c_1(1+\sqrt{2})^n + c_2(1-\sqrt{2})^n\;$형태가 나오며, $\;\;c_1, c_2$는 따로 계산한다.

\
- **Example**

　　　image.png

- **Assume**
  - $1\over 10\;$ of people outside $cA\;$ move in
  - $2\over 10\;$ of people inside $cA\;$ move out
  - No birth and No die

　image.png

- - $\begin{cases} y_{k+1} = 0.9y_k + 0.2z_k \\ z_{k+1} = 0.1y_k + 0.8z_k\end{cases} \quad\to\quad \begin{bmatrix} y_{k+1} \\ z_{k+1} \end{bmatrix} \;=\; \begin{bmatrix} 0.9&0.2 \\ 0.1&0.8 \end{bmatrix} \begin{bmatrix} y_k \\ z_k \end{bmatrix} \quad\to\quad$ 이때 행렬을 **Markov Matrix**라고 한다.

  - $u_{k+1} \;=\; Au_k\;$ 형태

  - $u_k \;=\; [e_1 \;\; e_2] \begin{bmatrix} \lambda_1^k&0 \\ 0&\lambda_2^k\end{bmatrix} \begin{bmatrix} c_1 \\ c_2 \end{bmatrix} \quad\quad from\;\; S^{-1}u_0 \;=\; [c_1\;\;c_2]^T \quad\to\quad \begin{bmatrix} y_k \\ z_k \end{bmatrix} \;=\; c_1\lambda_1^ke_1 \;+\; c_2\lambda_2^ke_2$

\
- **특징**
  - **Markov matrix의 column 합 = 1**
  
  - **Markov Eigenvalue의 꼭 하나는 $\;\lambda_1 \;=\;1 \;\;$나머지는 $\;|\lambda_i| \;<\;1$**

  - **$\lim_{\;k\to \infty} u_k \quad\to\quad$ (일반적으로) 수렴(coverge) = Stable state $\quad\to\quad u_s \;=\;Au_s$**

\
- **Solution**
  - $\begin{vmatrix} 0.9-\lambda&0.2 \\ 0.1&0.8-\lambda \end{vmatrix} \;=\; 0$

  - $\lambda^2 - 1.7\lambda +0.7 = 0$

  - $\lambda = 1 \quad or \quad 0.7$

  - Eigenvectors $\;S \;=\; \begin{bmatrix} 2&1 \\ 1&-1 \end{bmatrix}$

  - $\lambda = 1 \quad\to\quad \begin{bmatrix} -0.1&0.2 \\ 0.1&-0.2 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \quad\to\quad -0.1x_1+0.2x_2 = 0 \quad\to\quad eigenvector : \begin{bmatrix} 2 \\ 1 \end{bmatrix}$

## **IV. Differential & $e^{At}$**
---

### **1. Basic**

- $y(t)  \;=\; ce^{at} \quad-{d\over dt}\to\quad y'(t) \;=\; cae^{at} \;=\; ay(t) \quad\quad\quad c\;:\;Constant$

\
- $y'(t) \;=\; ay(t) \quad\to\quad y'(t) - ay(t) \;=\; 0 \quad\color{red}{\to}\quad$ **역으로 이런 관계로 만족시키는 $\;y(t) \;=\; e^{at}$이다.**

\
- $y'(t) - ay(t)  \;=\; 0 \quad : \quad$ **Homogeneous Ordinary 1st-order Differential Equation** 

\
- $y' \;=\; {dy \over dt} \quad\to\quad y' - ay \;=\;0 \quad\to\quad {dy \over dt} \;=\; ay \quad\to\quad {1 \over y} dy \;=\; a\;dt \quad\to\quad \int{1\over y} dy \;=\; \int a dt \quad\to\quad lny \;=\; at + \tilde{c} \quad\to\quad y(t) \;=\; e^{at} e^{\tilde{c}} \;=\; ce^{at} \;,\quad c\;=\; e^{\tilde{c}}$

\
- $\color{darkred}{y'(t) + ay(t) \;=\; 0\;이면,\;\;Solution은\;\;y(t) \;=\; e^{\lambda t}일\;것이며\;\;\lambda만\;\;구하면\;된다.}$

  - $\lambda e^{\lambda t} + ae^{\lambda t} \;=\; 0$
  - $\lambda + a \;=\; 0 \quad\because\;e^{\lambda t} \ne 0$
  - $\lambda \;=\; -a$
  - $y(t) \;=\; ce^{-at} \quad\to\quad$ **Basis function**

\
- **< 차분 방정식 >** $\quad a_{n+1} +ra_n \;=\;0 \;\to\; an_n \;=\;(\lambda)^n\;,\;\; \lambda+r \;=\;0\;\;$과 비슷한 방식이다.

  - $(\lambda)^n\;$이냐$\;e^{\lambda t}\;$이냐가 **Differnce eqn**과 **Differential eqn**의 차이

### **2. 확장**

- **$\color{darkred}{Difference\;\;equation\;\;(차분\;\;방정식)}$**

  - $pa_{n+2} + qa_{n+1} +ra_n \;=\; 0$
    
    - $a_n = c_1(\lambda_1)^n + c_2(\lambda_2)^n\quad$형태이며, $\quad \lambda_1 = \lambda_2 = \lambda$이면, $\\;\\; c_1(\lambda)^n + c_2n(\lambda)^n$이다.

  - $a_n = \lambda^n$으로 가정하면, $\;p\lambda^{n+2} + q\lambda^{n+1} + r\lambda^n \;=\;0$
    - $p\lambda^2 + q\lambda + r \\;=\\;0 \quad\to\quad Characteristic\\;eqn \quad\to\quad \lambda_1, \lambda_2$계산.

  
\
- **$\color{darkred}{Differential\;\;equation\;\;(미분\;\;방정식)}$**

  - $y'' + ay' + by \;=\; 0 \quad\to\quad$ **Homogeneous ordinary 2nd-order differential eqn**

  - $y(t) \;=\; e^{\lambda t}\;\;$형태일 것이므로 대입하면,

  - $\lambda^2e^{\lambda t} + a\lambda e^{\lambda t} + be^{\lambda t} \;=\;0\;$이고, 정리하면

  - $\lambda^2 + a\lambda + b \;=\;0 \quad\to\quad characteristic eqn\;\;$이다.

  - 위 식으로부터 $\;\lambda_1, \lambda_2$를 구해 해를 구한다.

  - $c_1e^{\lambda_1 t} + c_2e^{\lambda_2t}$
    - $If\quad \lambda_1 = \lambda_2 = \lambda\\;$이면, $\quad c_1e^{\lambda t} + c_2te^{\lambda t}\\;$일 것이며 이때는 $\\; a^2-4bc \\;=\\;0\\;$의 조건이 필요하다.

\
- $\color{darkred}{Matrix\;\;(행렬)}$

  - $u_{k+1} \;=\; Au_k \quad\to\quad u_k \;=\;A^ku_0 \;=\; S\Lambda^kS^{-1} \;=\; c_1\lambda_1^kx_1 + c_2\lambda_2^kx_2\quad,\; S^{-1}u_0 \;=\;\begin{bmatrix} c_1 \\ c_2 \end{bmatrix}\;$을 이용.

  - $u' \;=\; Au \quad\to\quad c_1e^{\lambda_1 t}x_1 + c_2e^{\lambda_2t}x_2\;$일 것이다.

  - $\Lambda^k = \begin{bmatrix} \lambda_1^k&0&\cdots&0 \\ 0&\lambda_2^k&\cdots&0 \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&\lambda_n^k \end{bmatrix} \quad\quad e^{t \Lambda} = \begin{bmatrix} e^{\lambda_1t}&0&\cdots&0 \\ 0&e^{\lambda_2t}&\cdots&0 \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&e^{\lambda_nt} \end{bmatrix}$

  - $u'(t) \;=\; \begin{bmatrix} x'(t) \\ y'(t) \end{bmatrix} \;=\; \begin{bmatrix} ax(t) + by(t) \\ cx(t) + dy(t) \end{bmatrix} \;=\; \begin{bmatrix} a&b \\ c&d \end{bmatrix}u$
    - 2행을 보면, $\\; x(t) \\;=\\; {1\over c}y'(t) - {d\over c}y(t)$이고, 이 식을 1행에 대입하면

    - ${1\over c}y''(t) - {d\over c}y'(t) \\;=\\; a({1\over c}y'(t) - {d\over c}y(t)) + by(t)$이다.

    - 정리하면 $\\; y''(t) - (a+d)y'(t) + (ad-bc)y(t) \\;=\\; 0$이고, $\\;y(t) = e^{\lambda t}$로 치환해 정리하면

    - $\lambda^2 - (a+d)\lambda + (ad-bc) \\;=\\; 0$이고, 이를 사용해

    - $\lambda_1, \lambda_2$를 구해

    - $y(t) \\;=\\; c_1e^{\lambda_1 t} + c_2e^{\lambda_2 t}\quad x(t) \\;=\\; c_3e^{\lambda_1 t} + c_4e^{\lambda_2 t}$를 구한다.

  \
  - 다른 방식으로 나타내면,
 
  - $|A- \lambda I| \;=\; 0$

  - $\begin{vmatrix} a-\lambda&b \\ c&d-\lambda \end{vmatrix} \;=\; (a-\lambda)(b-\lambda)-bc \;=\;0 \quad\to\quad \lambda^2 -(a+d)\lambda +(ad-bc) \;=\;0$
  
  - $u = \begin{bmatrix} x(t) \\ y(t)\end{bmatrix} = e^{\lambda_1t}\begin{bmatrix} c_3 \\ c_!\end{bmatrix} + e^{\lambda_2t} \begin{bmatrix} c_4 \\ c_2 \end{bmatrix} \;=\; \begin{bmatrix} c_3&c_4 \\ c_1&c_2 \end{bmatrix}\begin{bmatrix} e^{\lambda_1t} \\ e^{\lambda_2t} \end{bmatrix}$

### **3. General** 

  - $u' \;=\; Au \quad\quad y'(t) \;=\; ay \;\to\; y(t)\;=\;ce^{at}$

  - $u(t) \;=\; ce^{tA}$를 아래의 두 조건을 사용해 변화시키면
    - **Taylor Series** $\\; f(x) = f(0) + f'(0)x + {f''(0)x^2 \over 2!} + {f'''(0)x^3 \over 3!} + \cdots$
      - $e^x = 1 + x + {x^2 \over 2!} + {x^3 \over 3!} + \cdots$

  - $Se^{t\Lambda}S^{-1} \;=\; s^{tA}$

  - $e^{t\Lambda} = \begin{bmatrix} e^{\lambda_1t}&0&\cdots&0 \\ 0&e^{\lambda_2t}&\cdots&0 \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&e^{\lambda_nt} \end{bmatrix} \quad\to\quad \color{darkred}{Eigen\;value의\;\;개념\;\;이용\;가능}$

\
- $||X||^2 \;=\; |x_1|^2 + |x_2|^2 + \cdots + |x_n|^2 \;=\; x_1^*x_1 + x_2^*x_2 + \cdots + x_n^*x_n \;=\; (X^T)^*X$

    - If $\quad X^Hy\quad$ is **Real value**, $\quad X^Hy = y^HX$

  - $(AB)^H \;=\; B^HA^H$

  - $A^H \;=\; A \quad\to\quad Hermitian\;\;Matrix$
  - $\begin{cases} a_{ij} \;=\; a_{ij}^* & i \ne j \\ a_{ii} \;\to\; real \end{cases}$ 

\
- 전부 실수 이면, Hermitian $\;\to\;$ Symmetric

### **5. Properties of Hermitian or Symmetric Matrix**

\
1. **$X^HAX \;(=\; quadratic\;\;form)\;\;is\;\;real\;\;number$**

  - **proof)** $\;(X^HAX)^H \;=\; (X^HA^HX) \;=\; X^HAX \quad\because\;\; A\;=\; Hermitian$

  - **quadratic form** : 기하학적인 모형 표현에 사용, 이차식 표현에 사용\
  $\quad\quad\quad \begin{bmatrix} x&y \end{bmatrix} \begin{bmatrix} a&0 \\ 0&b \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} \;=\; ax^2 + by^2$

\
- $R \;=\; A^HA \quad\to\quad X^HRX \;=\; X^HA^HAX \;=\; (AX)^H(AX) \;=\; |AX|^2 > 0 \quad\to\quad Eigenvalue$와 연관
  - $A^HA \;=\; R\;\;(or\;\;A^T=A)$ 로 두면 여기서 얻어지는 Eigenvalues, Eigenvectors는 특이한 값을 갖는다.

\
2. **Every Eigenvalue is Real**
  - $AX \;=\; \lambda X$ 의 양변에 $\;X^H$ 를 곱하면

  - $X^HAX \;= \; \lambda X^HX \;=\; \lambda ||X||^2$ 이 되며, 정리하면

  - $\lambda \;=\; {X^HAX \over ||X||^2}$ 이다. 

  - 이를 **R**과 연결시키면

  - $Eigenvalue \;\;\; \lambda \;=\; {X^HRX \over ||X||^2} \;\geq\;0$이 된다.

\
3. **Eigenvectors are Orthogonomal**

  - $AX_1 \;=\; \lambda_1X_1 \quad , \quad AX_2 \;=\; \lambda_2X_2 \quad \lambda_1 \ne \lambda_2$ 라 하자. 이를 이용하여 식 2개를 다음과 같이 생성하자.

  - **식 (1)** $\quad (AX_1)^HX_2 \;=\; X_1^HA^HX_2 \;=\; X_1^HAX_2 \;=\; \lambda_2X_1^HX_2 \quad\because\;\; AX_2 = \lambda_2X_2$

  - **식 (2)** $\quad (AX_1)^HX_2 \;=\; (\lambda_1X_1)^HX_2 \;=\; \lambda_1^*X_1^HX_2 \;= \;\lambda_1X_1^HX_2 \quad\because\;\; \lambda_1 \;=\; real \;\;\therefore\;\; \lambda_1^* \;=\; \lambda_1$

  - **식(1) = 식(2)**이므로, $\quad (\lambda_1-\lambda_2)X_1^HX_2 \;=\;0$

  - $\color{darkred}{따라서}$, $\;\;\lambda_1 \ne \lambda_2$ 이므로, $X_1^HX_2 \;=\; 0$ 이므로  $X_1$ 과 $X_2$ 는 **Orthogonal** 이다.

### **6. Spectral Theorem**

- $Q \;:\; Orthonomal\;\;matrix$

  - $Q \;=\; \begin{bmatrix} q_1&q_2&\cdots&q_n \end{bmatrix} \quad\quad q_i^Tq_j = \begin{cases} 0&i\ne j \\ 1&i = j\end{cases}$

  - $Q^TQ \;=\; I \quad\to\quad Q^{-1} \;=\; Q^T$

\
- $A \;:\; Symmetric \quad \lambda_1, \lambda_2, \cdots, \lambda_n\quad X_1, X_2, \cdots X_n$

  - **가정** : $\;||X||^2 \;=\; 1$

  - $A \;=\; S\Lambda S^{-1}$는 $\quad Q\Lambda Q^{-1}$ 로 표현 가능.

  - $Q\Lambda Q^{-1} \;=\; \begin{bmatrix} x_1&x_2&\cdots&x_n \end{bmatrix} \begin{bmatrix} \lambda_1&0&\cdots&0 \\ 0&\lambda_2&\cdots&0 \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&\lambda_n \end{bmatrix} \begin{bmatrix} x_1^T \\ x_2^T \\ \vdots \\ x_n^T \end{bmatrix} \;=\; \begin{bmatrix} \lambda_1x_1&\lambda_2x_2&\cdots&\lambda_nx_n \end{bmatrix} \begin{bmatrix} x_1^T \\ x_2^T \\ \vdots \\ x_n^T \end{bmatrix}$

  - $\color{red}{\therefore}\quad A \;=\; \lambda_1X_1X_1^T + \lambda_2X_2X_2^T + \cdots + \lambda_nX_nX_n^T \quad\to\quad \color{red}{Spectral\;\;Theorem}$

    - **$\lambda_iX_iX_i^T \quad=\quad X_i$ 방향으로 proection을 의미**

    - $\lambda_1 > \lambda_2 > \cdots > \lambda_n \quad \lambda \uparrow\quad : \quad \color{darkred}{Principal\quad components}$

\
- $U \;=\; Unitary \;\; matrix$
  
  - $\to\;\; Complex \;\; matrix$

  - $U^HU \;=\; UU^H \;=\; I$

  - $U^H \;=\; U^{-1}$

  - **특징**
    - Angle and Length preserved
    
      - (Angle) $\quad (UX)^H(Uy) \\;=\\; X^Hy \quad\cdots\cdots (1)$

      - (Length) $\quad||UX||^2 \\;=\\; ||X||^2 \quad\cdots\cdots (2)$

    \
    - Unitary Matrix의 Eigenvalues $|\lambda_i| = 1$

      - $UX \quad=\quad \lambda X$

      - $||UX|| = ||\lambda X|| = |\lambda|\cdot||X||$

      - $||X|| = |\lambda|\cdot||X|| \quad\quad from\quad(1)$

      - $\therefore\quad |\lambda| = 1 \quad\cdots\cdots (3)$

    \
    - Eigenvectors are Orthogo-nomal

      - $UX_1 = \lambda_1X_1\quad, \quad UX_2 = \lambda_2 X_2 \quad,\quad \lambda_1 \ne \lambda_2$

      - $X_1^HX_2 = (UX_1)^H(UX_2) \quad\quad from\quad (2)$
      - $= \quad (\lambda_1X_1)^H(\lambda_2X_2)$
      - $= \quad \lambda_1^*\lambda_2X_1^HX_2$
      - $\therefore \quad X_1^HX_2 = \lambda_1^*\lambda_2X_1^HX_2$ 이므로 정리하면
      - $(1 - \lambda_1^*\lambda_2)X_1^HX_2 = 0$

      - $(1 - \lambda_1^*\lambda_2) \ne 0$ 임을 증명하면, $X_1^HX_2 = 0$이 된다.
      - $\lambda_1^* \lambda_1 = 1,\quad \lambda_2^* \lambda_2 = 1 \quad\quad from \quad (3)$ 이므로 $\quad \lambda_1^* \lambda_2 = 1$ 이려면 $\lambda_1 = \lambda_2$이어야 하므로 $\lambda_1^* \lambda_2 \ne 1$이다.
      - $\therefore\quad X_1^HX_2 = 0$ 이다.

      

      





## **VI. Singular Value Decomposition (SVD)**
---

### **1. Basic**

- $A_{n\times n} = S\Lambda S^{-1} \quad for\quad n \times n$ square matrix

\
- $A_{m \times n} = U\Sigma V^T \quad\to\quad \color{red}{Non-square}$ **matrix에 적용 가능!** 

　　　　image.png

### **2. Method**

- **First, Find the Eigenvalues & Eigenvectors of $A^TA_{n\times n} \quad\to\quad A^TA \;:\;$ square matrix로 만들어 사용하기 위해서**

  - $A^TA \quad\to\quad Symmetric$

  - $X^TA^TAX \;=\; \lambda X^TX \quad\to\quad \lambda = {||AX||^2 \over ||\lambda||^2} \;\geq\; 0 \quad\to\quad$ **Eigenvalues are non-zero**

\
- **Assume that there are "r" non-zero eigenvalues**

　　　　image.png

- $\sqrt{\lambda_i} \;=\; \sigma_i \quad (i\;=\; 1,2,\cdots,r)$
  - $\sigma_i \;=\; Singular\;\;value$

\
- $\Sigma_1 \;=\; \begin{bmatrix} \sigma_1&0&\cdots&0 \\ 0&\sigma_2&\cdots&0 \\ \vdots&\vdots&\ddots&\vdots \\ 0&0&\cdots&\sigma_r \end{bmatrix} \quad\to\quad \color{red}{\Sigma_{m\times n}} \;=\; \begin{bmatrix} \Sigma_1&O \\ O&O \end{bmatrix}$ 

\
- $\color{red}{V \;=\; \begin{bmatrix} V_1 & V_2 \end{bmatrix}}$

  - $V_1$
    - **Row space (basis)**

    - $V_1 = [v_1 \quad v_2 \quad \cdots \quad v_r] \quad \in \quad R^n$
    - $C(A^T) \quad Row-space$의 basis를 계산.

  \
  - $V_2$
    - **Null space (basis)**

    - $V_2 = [v_{r+1} \quad v_{r+2} \quad \cdots \quad v_n] \quad \in \quad R^n$

    - $for \quad \lambda_{r+1}=\lambda_{r+2}=\cdots=\lambda_{n}=0$
      
      - $A^TAX_j = \lambda_{j}X_j = 0\quad,\quad (j = r+1,r+2,\cdots,n)\quad\quad from\quad \lambda_j = 0$

      - $\therefore\quad A^TAX_j = 0 \quad\to\quad X_j= N(A^TA)$**로 $A^TA$ 의 Null space 이다.**

      - $N(A^TA) = N(A)$와 같다.

      - $v_{r+1} \quad v_{r+2} \quad \cdots \quad v_n \quad\to\quad N(A)\quad basis$

  \
  - $V^TV \;=\; I$

  \
  - $V \;=\; \begin{bmatrix} V_1 & V_2 \end{bmatrix} \;=\; \begin{bmatrix} Row\;space\;basis & Null\;space\;basis \end{bmatrix}$

- $\color{red}{A \;=\; U\Sigma V^T}$

  - $AV \;=\; U\Sigma \;=\; \begin{bmatrix} u_1&u_2&\cdots&u_r&u_{r+1}&\cdots&u_n \end{bmatrix}_{m\times m}^{\in R^m} \begin{bmatrix} \sigma_1&0&\cdots&\cdots&\cdots&0 \\ 0&\ddots&\vdots&\vdots&\vdots&\vdots \\ \vdots&\vdots&\sigma_r&\vdots&\vdots&\vdots \\ \vdots&\vdots&\vdots&0&\vdots&\vdots \\ \vdots&\vdots&\vdots&\vdots&\ddots&\vdots \\ 0&\cdots&\cdots&\cdots&\cdots&0  \end{bmatrix}_{m\times n}$

  - $A \begin{bmatrix} v_1&v_2&\cdots&v_r&v_{r+1}&\cdots&v_n \end{bmatrix} \;=\; \begin{bmatrix} Av_1&Av_2&\cdots&Av_r&Av_{r+1}&\cdots&Av_n \end{bmatrix} \;=\; \begin{bmatrix} \sigma_1u_1&\sigma_2u_2&\cdots&\sigma_ru_r&0&\cdots&0 \end{bmatrix}$

  - $Av_i \;=\; \sigma_iu_i$

  - $u_i \;=\; {1 \over \sigma_i}Av_i \quad \in R^m \quad (i \;=\; 1,2,\cdots, r)$

    - $u_i^Tu_j \\;=\\; {1 \over \sigma_i}(Av_i)^T {1 \over \sigma_j}Av_j \\;=\\; {1 \over \sigma_i \sigma_j}v_i^TA^TAv_j \\;=\\; {1 \over \sigma_i \sigma_j}v_i^T\lambda_jv_j \\;=\\; \delta_{ij}$

    - $\therefore u_i = Orthogo-nomal$
    
    - $\quad\to\quad C(A)  basis$
    - $\quad\to\quad U_1 = [u_1\quad u_2\quad \cdots\quad u_r]$

  - $u_{r+1}, u_{r+2}, \cdots, u_m \in R^m$

    - **Left Null space $N(A^T) \quad\to\quad Ortho-nomal\quad basis$를 구하면 된다.**

    - $U_2 = [u_{r+1}\quad u_{r+2}\quad \cdots u_m]$

  \
  - $\color{red}{U_{m\times m}} \;=\; [U_1 \quad U_2]$


\
- $\color{red}{A\;=\;U\Sigma V^T} \;=\; \begin{bmatrix} U_1&U_2 \end{bmatrix} \begin{bmatrix} \Sigma_1&0 \\ 0&0 \end{bmatrix} \begin{bmatrix} V_1^T \\ V_2^T \end{bmatrix}$



### **3. Example**

- $A \;=\; \begin{bmatrix} 1&11 \\ 1&1 \\ 0&0 \end{bmatrix} \quad\to\quad A^TA \;=\; \begin{bmatrix} 2&2 \\ 2&2 \end{bmatrix}$

\
- $Eigen\;value \quad:\quad \begin{vmatrix} 2-\lambda &2 \\ 2& 2-\lambda \end{vmatrix} \;=\; 0 \quad\to\quad \lambda_2 \;=\; 4, \;\; \lambda_2 \;=\; 0$

\
- $Eigen\;vector$
  - $\lambda_1 \;=\; 4 \quad \begin{bmatrix} -2&2 \\2&-2 \end{bmatrix}\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \quad\to\quad x_1-x_2 \;=\; 0 \quad\to\quad {1 \over \sqrt{2}} \begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix} \quad=\quad V_1 \quad for\quad non\;\;zero\;\;eigenvalue$

  - $\lambda_1 \;=\; 0 \quad \begin{bmatrix} 2&2 \\2&2 \end{bmatrix}\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} \quad\to\quad x_1+x_2 \;=\; 0 \quad\to\quad {1 \over \sqrt{2}} \begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix} \quad=\quad V_2 \quad for\quad non\;\;zero\;\;eigenvalue$
    
    -  ${1 \over \sqrt{2}} \\;=\\;$ **크기 1을 맞추기 위해**

  - $\color{red}{\therefore \;\; V} \;=\; {1 \over \sqrt{2}} \begin{bmatrix} 1&1 \\ 1&-1 \end{bmatrix}$


\
- $u_1 \;=\; {1 \over \sigma_1} Av_1 \;=\; {1 \over 2} \begin{bmatrix} 1&11 \\ 1&1 \\ 0&0 \end{bmatrix} {1 \over \sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix} \;=\; {1 \over \sqrt{2}} \begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix} \quad\cdots\cdots (1)$

\
- $u_2 \;=\; A^{-1} \quad\to\quad \begin{bmatrix} 1&1&0 \\ 1&1&0 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix} \;=\; \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix} \quad\to\quad x_1+x_2 \;=\; 0 \quad\to\quad u_2 \;=\; {1 \over \sqrt{2}} \begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix} \quad from$ **(1)과 Ortho-nomal**$\quad\cdots\cdots (2)$

\
- $u_3 \;=\; x_1+x_2$ **이면서 (1), (2)와 수직인 것**$\quad\to\quad u_3 \;=\; \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}$

\
- $\color{red}{\therefore \quad A \;=\;} \begin{bmatrix} {1 \over \sqrt{2}}&{1 \over \sqrt{2}}&0 \\ {1 \over \sqrt{2}}&-{1 \over \sqrt{2}}&0 \\ 0&0&1 \end{bmatrix} \begin{bmatrix} 2&0 \\ 0&0 \\ 0&0 \end{bmatrix} \begin{bmatrix} {1 \over \sqrt{2}}&{1 \over \sqrt{2}} \\ {1 \over \sqrt{2}}& -{1 \over \sqrt{2}} \end{bmatrix}$

\
- $\color{red}{Cf) \quad SVD}\;:\;$ **Unique하지 않다. Only $\;\Sigma$만 unique 하다.**
